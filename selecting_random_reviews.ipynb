```python
import pandas as pd
import numpy as np
import random
```

```python
## loading in random rows method from https://stackoverflow.com/questions/22258491/read-a-small-random-sample-from-a-big-csv-file-into-a-python-data-frame
filename = "yelp_dataset\yelp_academic_dataset_review.csv"
n = sum(1 for line in open(filename, encoding = 'utf-8')) - 1 #number of records in file (excludes header)
s = 30000 #desired sample size
skip = sorted(random.sample(range(1,n+1),n-s)) #the 0-indexed header will not be included in the skip list
```

```python
raw_data = pd.read_csv(filename, skiprows=skip, on_bad_lines = 'skip')

raw_data.head(5)
```

```python
raw_data['user_id'] = raw_data['user_id'].str[2:-1]
raw_data['review_id'] = raw_data['review_id'].str[2:-1]
raw_data['text'] = raw_data['text'].str[2:-1]
raw_data['date'] = raw_data['date'].str[2:-1]
raw_data['business_id'] = raw_data['business_id'].str[2:-1]

raw_data.head(5)
```

```python
raw_data.to_csv('selected_yelp_reviews.csv', index=False)
```
